```python
import PIL
from PIL import Image
image=Image.open("hills.jpg")
```

```python
#Now, lets try to crop the image and store it in another variable (cropped_image).
co_ordinates=(0,0,500,417)
cropped_image=image.crop(co_ordinates)
display(cropped_image,image)
```

```python
#We have initialised a tuple named co_ordinates that represent the following.
#(left, top, right, bottom)
#The same image is shown in the form of a graph for better understanding.
import matplotlib.pyplot as plt
plt.imshow(cropped_image)
print(cropped_image)
#Play with the values for better understanding.
#NOTE: matplotlib is a plotting library and not a part of PIL.
```

```python
#There are two types in flipping an image, horizontally and vertically.
#Four types of image rotations are possible in PIL.
#Now, let us see each and every one of them.
horizontal_flip=image.transpose(Image.FLIP_LEFT_RIGHT)
vertical_flip=image.transpose(Image.FLIP_TOP_BOTTOM)
rotate=image.transpose(Image.ROTATE_90)
transpose=image.transpose(Image.TRANSPOSE)
display(horizontal_flip,vertical_flip,rotate,transpose,image)
```

```python
#The other three rotations can be obtained by changing the values (90,180,270,360) in the place of
#ROTATE_xx
#Image.TRANSPOSE transposes the image.
```

```python
#Now lets try resizing the image (either increase or decrease). Let us try both.
#Keep in mind the original dimensions (626, 417).
larger_image=image.resize((700,500))
smaller_image=image.resize((500,300))
display(larger_image,smaller_image,image)
#Here, we have increased and decreased the size of original 'hills' image, but the original aspect ratio
#has changed. Which means that the original image and the other two images don't have the 
#same aspect ratio. But they are resized irregularly.
```

```python
#Inorder, to preserve the aspect ratio, use this,
from PIL import ImageDraw
same_aspect_ratio=image.copy()
same_aspect_ratio.thumbnail((500,300))
display(same_aspect_ratio,image)
print(same_aspect_ratio.size,image.size)
#The thumnbail preserves the aspect ratio of the original image. But this method can't yield 
#an image larger than the original.
#Divide the width and height of both images, they'll be equal.
```
