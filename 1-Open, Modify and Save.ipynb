```python
#Importing the library
#The "import PIL" command imports the Imaging Library into Python.
import PIL
```

```python
#In order to open/display/modify an image we need to import the "Image" module from PIL.
from PIL import Image
```

```python
#Great, now we're ready to go!
#To open an image, the respective image must be placed in 
#"C:\Users\(The one folder other than 'Public' and 'Default')\"
#Use the Image.open() command to open an image.
image=Image.open("hills.jpg")
```

```python
#Now that we have successfully opened the image and stored it in the variable named 'image',
#we shall now proceed to display the image.
display(image)
```

```python
#That's how simple it is!
#Now, if you notice keenly, we have used the keyword 'display', instead of 'print'.
#What happens if we use print()?
print(image)
```

```python
#As you can see, the print prints (text) some of the basic properties like the mode='RGB', size, etc,.
#There are other ways to display an image.
image.show()
#This command opens a dialog box, prompting you to choose an image viewer in your PC.
#Choose whatever you like. This may take some time for first timers and then it'll be instantaneous.
```

```python
#Let us print some of the image properties in a more readable manner.
print(image.size)
print(image.format)
print(image.mode)
#image.size prints the (width, height) of an image
#image.format prints the format. Here 'JPEG'. Some other formats are 'png','GIF' etc,.
#image.mode prints the mode. Like 'RGB', 'RGBA', 'BGR' etc,.
```

```python
#Saving, copying and changing the format.
#Let us create a new variable 'image_copy' and copy the original image in it.
image_copy=image.copy()
display(image_copy)
#Now a copy of the original "hills" image gets copied to the 'image_copy' variable.
```

```python
#Now, if you need the copy of the orignal "hills" image to be stored in your PC, use,
image_copy.save("hills_copy.jpg")
#You can enter whatever name you want inside the braces. Just don't forget the format.
#Go to "C:\Users\(The one folder other than 'Public' and 'Default')\" folder to view it.
```

```python
#You can also change the format of the image by simply changing the extension in save()
image_copy.save("hills_copy_png.png")
```
